# Exercici 2

Trobeu i classifiqueu els punts crítics de
$$f(x, y) = 2x^4 + y^4 − x^2 − 2y^2$$
Representeu gràficament $f$ i les seves corves de nivell.
Si cal, feu més d'un gràfic amb l'objectiu que es vegi clarament el comportament a cada punt crític.

```python
f(x,y)=2*(x^4) + y^4 - x^2 - 2*(y^2)
eqx = f.diff(x)==0
eqy = f.diff(y)==0
llista = solve([eqx,eqy],(x,y))
for i in range (len(llista)):
    hess_f=f.diff().diff()
    show(hess_f.subs(llista[i]))
    hessiana = hess_f.subs(llista[i])
    if det(hessiana) > 0:
        if hessiana[0,0] > 0:
            print("Matriu definida positiva, el Punt:", str(llista[i]), "és un mínim local")
        if hessiana[0,0] < 0:
            print("Matriu definida negativa, el Punt:", str(llista[i]), "és un màxim local")
    else:
        print("El Punt:", str(llista[i]), "és un punt de sella")

A = plot3d(f(x,y),(x,-1,1),(y,-1,1))
#B = contour_plot(f(x,y), (x,-1,1), (y,-1,1), plot_points=250, contours=15,colorbar='True')
show(A)
contour_plot(f(x,y), (x,-1,1), (y,-1,1), plot_points=250, contours=15,
colorbar='True')
```
